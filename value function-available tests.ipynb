```python
import numpy as np
from scipy.stats import poisson
from scipy.stats import dirichlet
from scipy.stats import beta
import matplotlib.pyplot as plt


n=5000    #number of particles#
s=[]

for i in range(0,n):
    sta=[]
    sta.append(np.random.beta(1,20))
    theta=np.ones(2)
    theta[0]=np.random.beta(2,6)
    theta[1]=np.random.beta(1,9)
    sta.append(theta)
    sta.append(5)
    s.append(sta)
    
Belief1=np.zeros(n)
Belief2=np.zeros(n)
Belief3=np.zeros(n)
Belief4=np.zeros(n)
Belief5=np.zeros(n)
Belief6=np.zeros(n)
Belief7=np.zeros(n)
Belief8=np.zeros(n)
Belief9=np.zeros(n)
Belief10=np.zeros(n)
for i in range(n):
    Belief1[i]=beta.pdf(s[i][0],1,20)*beta.pdf(s[i][1][0],2,6)*beta.pdf(s[i][1][1],1,9)
    Belief2[i]=beta.pdf(s[i][0],3,10)*beta.pdf(s[i][1][0],10,6)*beta.pdf(s[i][1][1],3,9)
    Belief3[i]=beta.pdf(s[i][0],6,8)*beta.pdf(s[i][1][0],9,1)*beta.pdf(s[i][1][1],10,11)
    Belief4[i]=beta.pdf(s[i][0],10,6)*beta.pdf(s[i][1][0],5,19)*beta.pdf(s[i][1][1],1,11)
prior1=Belief1/np.sum(Belief1) 
prior2=Belief2/np.sum(Belief2) 
prior3=Belief3/np.sum(Belief3) 
prior4=Belief4/np.sum(Belief4) 
prior5=0.3*prior1+0.7*prior2
prior6=0.9*prior1+0.1*prior3
prior7=0.5*prior1+0.5*prior4
prior8=0.45*prior2+0.55*prior3
prior9=0.66*prior2+0.34*prior4
prior10=0.5*prior3+0.5*prior4

```

```python
import SIS
import POMCP
V=[]
T=50
budget=float("inf")
max_play=50

    
```

```python
sis=SIS.SIS(s,prior1)
belief=sis.belief
total_rewards=0
pomcp=POMCP.POMCP(sis)
pomcp.add_configs(budget, belief, reinvigorated_particles_ratio=0.1, utility_fn='ucb1', C=0.5)
a=[]
bel=[]
#traj=[]
#traj.append(sis.curr_state)
for i in range(max_play):
            # plan, take action and receive environment feedbacks
        pomcp.solve(T)
        action = pomcp.get_action(prior1)
        print(action)
        a.append(action)
        new_state, obs, reward, cost = pomcp.model.take_action(action)
        state=new_state
        #traj.append(new_state)
          # update states
        belief = pomcp.update_belief(belief,action, obs)
        #bel.append(belief)
        total_rewards += reward
        budget -= cost
 
print(total_rewards)
```
